```python
from itertools import combinations
# Agrupar por 'ClientID' y obtener listas de agentes por cliente
client_agents = archivo.groupby('ClientID')['AGENT.ID'].apply(list)

# Crear un diccionario para contar combinaciones de agentes
agent_pairs = {}

# Iterar sobre cada lista de agentes y crear combinaciones
for agents in client_agents:
    for pair in combinations(sorted(agents), 2):
        if pair in agent_pairs:
            agent_pairs[pair] += 1
        else:
            agent_pairs[pair] = 1

# Convertir el diccionario en un DataFrame
Agentes = pd.DataFrame([(k[0], k[1], v) for k, v in agent_pairs.items()],
                         columns=['Agente1', 'Agente2', 'ClientesComun'])

Agentes
```

```python
Agentes = Agentes.loc[Agentes['Agente1']!=Agentes['Agente2']]
```

```python
Agentes = Agentes.sort_values(by='ClientesComun', ascending=False).reset_index(drop=True)
Agentes
```

```python
Agentes.to_csv('Relacion de Agentes.csv')
```
